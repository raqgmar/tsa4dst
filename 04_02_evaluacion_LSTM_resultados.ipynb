# 2. Visualización de las predicciones del mejor modelo.

- Graph the Storms (Dst vs. Datetime) Comparison

```python
def plot_storm_comparison(storm_data, storm_name):
    plt.figure(figsize=(12, 6))
    plt.plot(storm_data['Datetime_pred'], storm_data['y_true'], label='Real Dst')
    plt.plot(storm_data['Datetime_pred'], storm_data['y_pred'], label='Predicted Dst')
    plt.xlabel('Datetime')
    plt.ylabel('Dst')
    plt.title(f'Storm Comparison: Real vs. Predicted Dst ({storm_name})')
    plt.legend()
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='Dst_real', y='Dst_pred', data=storm_data)
    plt.xlabel('Real Dst')
    plt.ylabel('Predicted Dst')
    plt.title(f'Scatter Plot: Real vs. Predicted Dst ({storm_name})')
    plt.show()
```
