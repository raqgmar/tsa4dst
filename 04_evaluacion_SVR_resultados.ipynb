## 2. Exploración de las iteraciones realizadas sobre el grid.

```python
def show_parameters_grid_executions(path_resultados,model_path):
    # Conectar a la base de datos
    conn = sqlite3.connect(path_resultados+model_path)

    # Extraer los parámetros de los ensayos
    trial_params = pd.read_sql_query("SELECT * FROM trial_params;", conn)

    # Pivotear los datos para ver cada ensayo con sus parámetros
    params_pivot = trial_params.pivot(index='trial_id', columns='param_name', values='param_value')

    # Cerrar la conexión
    conn.close()

    # Mostrar las combinaciones únicas de parámetros
    unique_combinations = params_pivot.drop_duplicates()
    unique_combinations_count = unique_combinations.shape[0]

    # Crear el pairplot
    sns.pairplot(unique_combinations)
    plt.show()

    unique_combinations_count, unique_combinations
```
